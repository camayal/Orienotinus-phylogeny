## Different tests with multiple alpha or beta
Those parameters are imporant in the network reconstruction, and calibrate those are not trivial. A Simplex plot should no have green or yellow elements and green elements must be: 1) in the three vertices (for tree reconstructions) or 2) along central lines that connect all vertices (for msc reconstructions as astral). Check this video for a further explaination: https://www.youtube.com/watch?v=KLFldo1JSr0

```r
#Modify distributions base on new alphas and betas
#In this case NANUQ() function is not use anymore (it is too slow), instead NANUQdist() can modify our distributions
#based on the new alpha and beta.
alpha = 0.01
beta = 0.95
nanuqDist_sec = NANUQdist(nanuqDist, alpha = alpha, beta = beta,
                    outfile = paste("/home/camayal/viburnumThings/Viburnum-Oreinotinus/notebooks/Oct2020/analysis-nanuq/NANUQdist_a",alpha,"b",beta),
                    plot = TRUE)
```

### Check multiple posibilities using a programatic way

```r
#Test multiple values of alpha and beta, and save both, matrix in nexus and simplex plot in jpg.

#before doing this nanuqDist object must be created using NANUQ() function (check first cells)

folder = "/home/camayal/viburnumThings/Viburnum-Oreinotinus/notebooks/Oct2020/analysis-nanuq/MultipleABfig_2/"


dir.create(paste(folder,"simplex", sep=""))

count = 0
for (a in 0:10) {
alpha = ((0.1 ^ a)-(0.1 ^ a)/20)
    for (b in 0:10) {
        count = count + 1
        beta = ((0.1 ^ b)-(0.1 ^ b)/20)
        jpeg(paste(folder,"simplex/",formatC(count, width=3, flag="0"),"_NANUQdist_alpha",alpha,"_beta",beta,".jpg", sep=""))

        nanuqDist_sec = NANUQdist(nanuqDist, alpha = alpha, beta = beta,
                            outfile = paste(folder,formatC(count, width=3, flag="0"),"_NANUQdist", sep=""),
                            plot = TRUE)
        dev.off()
    }
}

```
